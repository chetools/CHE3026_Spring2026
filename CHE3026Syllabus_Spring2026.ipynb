##Topics

This is course focuses on theoretical model development and computational methods for heat and mass transfer<br>

1.  Introduction to momentum/heat/mass transfer analogies
2.  Heat diffusion equation, boundary and initial conditions
3.  1D conduction (plane wall and radial systems)
4.  2D and 3D conduction
5.  Steady and non-steady state finite difference methods, von Neuman stability criterion
6.  Convective heat transfer
7.  Heat exchangers - effectiveness/NTU
8.  Radiation heat transfer

9.  Fundamentals of mass transfer - mass/molar velocities and fluxes
10. Maxwell-Stefan relations
11. Diffusion in non-ideal mixtures - NRTL/thermodynamic factor matrix
12. Simplified cases - Ficks law, diffusion of A through stagnant B/equimolar counter-diffusion
13. Dimensional analysis - Buckingham pi
14. Interphase mass transfer
15. Absorption/stripping
