```python
import numpy as np
import scipy as sp
```

```python
A = np.array([[0, 0, 0, 1, 1, 1],
              [1, 1, 1, -3, -1, -1],
              [0, 0, -1, 0, -1, -2.]])
```

```python
ns=sp.linalg.null_space(A)
ns
```

```python
n1,n2,n3=ns[:,0], ns[:,1], ns[:,2]
```

```python
z = 20.*n1 + 50*n2 + 2*n3
z
```

```python
A@z
```

```python
##ns[(0,4,5),:] @ c1 = np.array([1, 0, 0.])
```

```python
c1 = np.linalg.solve(ns[(0,4,5),:], np.array([1, 0, 0.]))
c1
```

```python
ns @ c1  #pi1 = L/D
```

```python
c2 = np.linalg.solve(ns[(0,4,5),:], np.array([0, 1, 0.]))
c2
```

```python
ns @ c2 # mu/(rho V D)
```

```python
c3 = np.linalg.solve(ns[(0,4,5),:], np.array([0, 0, 1.]))
c3
```

```python
ns@c3 # deltaP/(rho*v**2)
```
