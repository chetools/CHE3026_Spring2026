```python
import numpy as np
import pandas as pd
df = pd.read_excel("PipeFlowData.xlsx", sheet_name='Sheet1')
```

```python
df
```

```python
LD = (df['L']/df['D']).values
Re = (df['rho']*df['V']*df['D']/df['mu']).values
dPrhov2 = (df['dP']/df['rho']/df['V']**2).values
```

```python
import plotly.express as px

fig = px.scatter_3d(x=LD, y=Re, z=dPrhov2, labels={'x': 'LD', 'y': 'Re', 'z': 'dPrhov2'})
fig.update_layout(title='3D Scatter Plot of LD, Re, and dPrhov2', scene_xaxis_type='log', scene_yaxis_type='log', scene_zaxis_type='log')
fig.show()
```

```python
A = np.c_[np.ones_like(Re), np.log(Re), np.log(LD)]
```

```python
A
```

```python
A.T @ A
```

```python
A.T @ np.log(dPrhov2)
```

```python
np.linalg.solve(A.T @ A, A.T @ np.log(dPrhov2))
```
